```python
cpi_df = pd.DataFrame()
cpi_df['date'] = [int(year) for year in date.split()]
```

```python
country_names = ['canada','france','germany','italy','japan','uk','us']
```

```python
def string_to_float(list_of_columns, list_of_names):
    for country, name in zip(list_of_columns,list_of_names):
        cpi_df[name] = [float(number) for number in country.split()]
```

#### Problem 1.1

```python
def calc_inflation(previous, current):
    return ((current - previous) / previous) * 100
```

```python
def get_inflation_values(country):
    inflation_list = []
    for i in range(1, len(cpi_df.iloc[:,country])):
        inflation_list.append(calc_inflation(cpi_df.iloc[i-1,country], cpi_df.iloc[i,country]))
    return inflation_list
```

```python
def get_inflation_df(country_list):
    inflation_df = pd.DataFrame()
    inflation_df['date'] = [int(year) for year in np.linspace(1973, 1997, 24)]
    for i in range(1, cpi_df.shape[1]):
        inflation_list = get_inflation_values(i)
        inflation_df[country_list[i-1]] = inflation_list
    return inflation_df
```

```python
inflation_df = get_inflation_df(country_names)
```
