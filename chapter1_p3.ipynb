```python
foreign_exchange_df = pd.DataFrame()
```

```python
foreign_exchange_df['date'] = [int(year) for year in date.split()]
```

```python
country_names = ['canada', 'france', 'germany', 'japan', 'sweden', 'switzerland', 'uk']
```

```python
def string_to_float(list_of_columns, list_of_names):
    for country, name in zip(list_of_columns,list_of_names):
        foreign_exchange_df[name] = [float(number) for number in country.split()]
```

```python
string_to_float([canada, france, germany, japan, sweden, switzerland, uk], country_names)
```

```python
fig, ax = plt.subplots(7,1, figsize=(15,20))
for i, name in zip(range(0,7), country_names):
    ax[i].plot(foreign_exchange_df.iloc[:,0], foreign_exchange_df.iloc[:, i+1])
    ax[i].set_title(name)
plt.subplots_adjust(hspace=0.75)
```
