```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from econtools import raw_to_df
```

```python
raw_data = """1967 463 468 466 514 467 492
1968 464 466 466 512 470 492
1969 459 466 463 513 470 493
1970 459 461 460 509 465 488
1971 454 457 455 507 466 488
1972 454 452 453 505 461 484
1973 446 443 445 502 460 481
1974 447 442 444 501 459 480
1975 437 431 434 495 449 472
1976 433 430 431 497 446 472
1977 431 427 429 497 445 470
1978 433 425 429 494 444 468
1979 431 423 427 493 443 467
1980 428 420 424 491 443 466
1981 430 418 424 492 443 466
1982 431 421 426 493 443 467
1983 430 420 425 493 445 468
1984 433 420 426 495 449 471
1985 437 425 431 499 452 475
1986 437 426 431 501 451 475
1987 435 425 430 500 453 476
1988 435 422 428 498 455 476
1989 434 421 427 500 454 476
1990 429 419 424 499 455 476"""
```

```python
column_names = ['year', 'verbal_male', 'verbal_female', 'verbal_total', 'math_male', 'math_female', 'math_total']
```

```python
sat_df = raw_to_df(raw_data.split(), column_names)
```

```python
sat_df
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,8))
ax[0].plot(sat_df.year, sat_df.verbal_male)
ax[0].plot(sat_df.year, sat_df.math_male)
ax[0].legend()
ax[1].plot(sat_df.year, sat_df.verbal_female)
ax[1].plot(sat_df.year, sat_df.math_female)
ax[1].legend()
```

```python
fig, ax1 = plt.subplots(figsize=(10,8))
ax1.plot(sat_df.year, sat_df.verbal_male)
ax2 = ax1.twinx()
ax2.scatter(sat_df.year, sat_df.verbal_female, color='r')
```

```python
fig, ax1 = plt.subplots(figsize=(10,8))
ax1.plot(sat_df.year, sat_df.math_male)
ax2 = ax1.twinx()
ax2.scatter(sat_df.year, sat_df.math_female, color='r')
```

```python
model = LinearRegression()
```

```python
fitted_model = model.fit(sat_df.verbal_male.values.reshape(-1,1), sat_df.verbal_female)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(sat_df.verbal_male, sat_df.verbal_female)
ax.plot(sat_df.verbal_male, fitted_model.predict(sat_df.verbal_male.values.reshape(-1,1)))
```
